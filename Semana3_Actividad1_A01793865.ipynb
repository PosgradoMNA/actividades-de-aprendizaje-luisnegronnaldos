**Fundamentos de Base de Datos y Para Ciencia de Datos**

Como lo indican las lecturas recomendadas, las bases de datos son grandes almacenes de información que permiten registrar diferentes tipos de eventos, en diferentes formatos e inclusive en diferentes tiempos (algunos datos se actualizan en el tiempo). 
 
El proceso tradicional de ingreso de información a la datawarehouse es el ETL, pero bajo las nuevas prácticas de subir información no estructurada a la nube, sobre todo a bases de datos tipo datalake se hace primera la extracción y carga de data y luego se genera el proceso de transformación según sea necesario. 
 
El uso de base de datos de gran tamaño se ha proliferado en los últimos años lo que ha dado espacio para el crecimiento ecosistemas de trabajo en big data, muchos de ellos en entornos de servidores en la nube que prometen la flexibilidad y velocidad requerida por los diferentes proyectos que ha permitido a entornos como Hadoop por ejemplo.
 
Un punto adicional a considerar es la veracidad de los datos, al tener una gran de información la probabilidad que se filtre información erronea es alta, por lo que se deben procesos de validación y consistencia de los datoa para garantizar que posteriormente puede ser usado proceso de machine learning por ejemplo.

En el caso de los Datawarehouse para proyectos de analítica se suele preferir bases de datos estructuradas, pero la cantidad de información actualmente hace que la preferencia por base de datos no estructuradas sean cada vez mayor, en este sentido las bases de datos denominadas Data Lake son una solución cada vez más usada, en mi experiencia profesional he trabajado con data lakes de AWS y estos han permitido mucha flexibilidad en el manejo de la información pero ha denotado un trabajo importante de transformación y ordenamiento antes de poder usarla en los proyectos de analítica de la empresa.

**Segunda Alternativa Reemplazando Valores **

```python
#Aqui se carga la base de datos desde el contenedor github. 
url = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
df = pd.read_csv(url)
df
```

```python
ndf = df.copy() #hacemos una copia de la base de datos
```

```python
Ximean = ndf.X2.mean() # Por ejemplo en este caso si uso la media, obtenemos valores que forman parte de la base de datos (los datos son ordinales)
Ximean
```

## Reemplazo de X2, X3 y X4 por el valor de la moda 

```python
Ximode = ndf.X2.mode() # En esta caso es mejor usa la mediana que me permite obtener un valor real para la base de datos 
Ximode
```

```python
ndf['X2'].fillna(value = Ximode, 
                    inplace = True)# Aqui estoy reemplazando los valores faltanes de X2 por la moda al ser un valor categorico 
```

```python
Ximode = ndf.X3.mode()
ndf['X3'].fillna(value = Ximode, 
                    inplace = True)# Aqui estoy reemplazando los valores faltanes de X3 por la moda al ser un valor categorico 
```

```python
Ximode = ndf.X4.mode()
ndf['X4'].fillna(value = Ximode, 
                    inplace = True)# Aqui estoy reemplazando los valores faltanes de X4 por la moda al ser un valor categorico 
```

## Reemplazo de X5 por el valor de la media

```python
Ximean = ndf.X5.mean()
ndf['X5'].fillna(value = Ximean, 
                    inplace = True)# Aqui estoy reemplazando los valores faltanes de X5 por la media 
```

## Eliminación de los datos faltantes para  de X6 a X11

```python
ndf.dropna(subset=['X6', 'X7', 'X8', 'X9', 'X10', 'X11'], inplace = True)
ndf # aqui elimino los datos faltantes de X5 a X11
```

```python
#import piplite
#import micropip
import pandas as pd
import numpy as np
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
import pandas as pd # Se ha cargado la data del caso del laboratorio para trabajar los ejercicios

df = pd.read_csv(path, header=None)
```

```python
df # Se comprueba que la base de datos ha cargado bien
```

```python
df.head(5) #con esto se puede mostrar tantas lineas como indique el valor en este caso las primeras cinco filas 
```

```python
df.tail(10) #con este comando se muestran las 10 ultimas final 
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers) # Se esta creando una lista para posteriromente asignarlo a la cabecera de cada columna. 
```

```python
df.columns = headers # con este comando lo asignamos a la cabecera de cada columna. 
df.head(10)
```

```python
df1=df.replace('?',np.NaN) # esta linea nos permite reemplzar el signo de interrogación por la nomenclatura de sin valor para que luego pueda ser limpiada 
```

```python
df=df1.dropna(subset=["price"], axis=0)
df.head(20)
```

```python
print(df.columns)
```

```python
#Comando para grabar la base de datos trabajada:

df.to_csv("automobile.csv", index=False) # Sin encabezado 
```

```python
df.to_csv("automobile1.csv", index=True) # Con encabezado
```

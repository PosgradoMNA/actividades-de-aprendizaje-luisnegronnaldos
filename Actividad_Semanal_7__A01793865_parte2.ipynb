**Definición de Latitud y Longitud**

**Latitud** Es la distancia en grados, minutos y segundos que hay con respecto al paralelo principal, que es el ecuador (0º). La latitud puede ser norte y sur.

**Longitud**: Es la distancia en grados, minutos y segundos que hay con respecto al meridiano principal, que es el meridiano de Greenwich (0º).La longitud puede ser este y oeste.

Ahora sí, implementa K means a los datos de latitud y longitud :) y encuentra donde colocar los almacenes. 

Nota: si te llama la atención implementar alguna otra visualización con otra librería, lo puedes hacer, no hay restricciones.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(latlong).score(latlong) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

# Definición de Clusters

# **Con 5 Clusters**


```python
kmeans = KMeans(n_clusters=5
                ).fit(latlong)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
asignar=[]
C = kmeans.cluster_centers_
labels = kmeans.predict(latlong)
colores=['red','green','blue','cyan','yellow']
for row in labels:
    asignar.append(colores[row])
f1 = df['latitude'].values
f2 = df['longitude'].values
 
plt.scatter(f1, f2, c=asignar, s=20)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.show()
```

# **Con 6 Clusters**


```python
kmeans = KMeans(n_clusters=6
                ).fit(latlong)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
asignar=[]
C = kmeans.cluster_centers_
labels = kmeans.predict(latlong)
colores=['red','green','blue','cyan','yellow','black']
for row in labels:
    asignar.append(colores[row])
f1 = df['latitude'].values
f2 = df['longitude'].values
 
plt.scatter(f1, f2, c=asignar, s=20)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.show()
```

# **Con 7 Clusters**


```python
kmeans = KMeans(n_clusters=7
                ).fit(latlong)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
asignar=[]
C = kmeans.cluster_centers_
labels = kmeans.predict(latlong)
colores=['red','green','blue','cyan','yellow','black','orange']
for row in labels:
    asignar.append(colores[row])
f1 = df['latitude'].values
f2 = df['longitude'].values
 
plt.scatter(f1, f2, c=asignar, s=20)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.show()
```

# Definiendo la Ubicación de las Tiendas

**Cantidad de Almacenes**

Se uso la curve 'Elbow Curve' para determinar una primera aproximación de cuantos centros de gravedad se podría explorar y se vio que entre 5 a 10 centros de gravedad sería adecuado. 

Se realizo la prueba con 5 a 7 centro de gravedad y se eligio 7 centros que generaba buenos puntos de distribución sin generar una gran cantida de centros. 

Para las zonas lejanas solo se considerao una que tenía los suficientes centros como para no generar sobre costos de tener centros de distribución para zonas cercanas. 

```python
#Se ubican los centros de las zonas más cercanas.
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, latlong)
closest
```

```python
#Se determinan las ciudades donde se realizarán los centros de distribución. 
users=df['name'].values
for row in closest:
    print(users[row])
```
